## Step. 4 
### Discover what is the mean age per occupation

```python
users[["age", "occupation"]].groupby("occupation").mean()
```

## Step. 5 
### Discover the Male ratio per occupation and sort it from the most to the least

```python
def gender_to_num(x):
    if x == 'F':
        return 0
    if x == 'M':
        return 1
    
users['gender_count'] = users['gender'].apply(gender_to_num)

male_ratio_per_occupation = users.groupby('occupation').gender_count.sum() / users.occupation.value_counts() 

male_ratio_per_occupation.sort_values(ascending = False)
```

## Step. 6
### For each occupation, calculate the minimum and maximum ages

```python
users.groupby('occupation').age.agg(['min', 'max'])
```

## Step. 7
### For each combination of occupation and sex and calculate the mean age

```python
users.groupby(['occupation', 'gender']).age.mean()
```

## Step. 8
### For each occupation present the percentage of women and men

```python
gender_occupation_count = users.groupby(['occupation', 'gender']).agg({'gender': 'count'})
occupation_count = users.groupby(['occupation']).agg('count')
occupation_gender_percent = gender_occupation_count.div(occupation_count, level = "occupation") * 100
occupation_gender_percent.loc[: , 'gender']
```

## Step. 4
### Year 2061? Do we really have data from this year? Create a function to fix it and apply it.

```python
def fix_century(x):
    year = x.year - 100 if x.year > 1989 else x.year
    return datetime.date(year, x.month, x.day)
wind['Yr_Mo_Dy'] = wind['Yr_Mo_Dy'].apply(fix_century)
wind.head()
```
